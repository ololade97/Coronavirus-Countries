```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.core.display import HTML

address = "https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/"
get_address = requests.get(address)
soup = BeautifulSoup(get_address.content, "lxml")
result = soup.find("table", {"id":"table3"})

df = pd.read_html(str(result))
df[0].to_csv("Coronavirus_Countries.csv")

d = pd.read_csv("Coronavirus_Countries.csv", index_col = 0)
display(HTML(d.to_html()))

```
